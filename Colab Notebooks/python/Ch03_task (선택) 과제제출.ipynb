<span style="font-size:80%">[문제5]<br><br></span>
```print( )문의 기능 및 특징에 대해 설명해주세요```

<span style="color:red">답변</span>

````python
a = 'hello'
b = 'is'
c = 'My name'
d = 'OZ coding'
e = 'I love'
f = '!'
print(a)
print(a , c , b , d)
print(e , d , f)

print(d, a, sep=",")
````

*   다양한 데이터들을 출력할 수 있음.
*   변수를 지정하여 변수들의 해당값을 출력할 수 있음.
*   여러 데이터를 ','로 구분하여 동시에 출력할 수 있음 (공백 삽입됨)
*   출력값 마다 구분자를 넣어 출력할 수 있음
