## for 문이란?
* 프로그래밍 언어에서 반복되는 작업을 간단하게 처리하기 위해서 사용하는 기능
* 콜론, 들여쓰기 중요함


```python
#2. for문으로 만들어보기
# for 변수 in range(횟수):
#   print()
```

```python
#3. for문 만들어보기
for i in range(5):
    print("반복이 되네 신기해!")
```

```python
#4. for문의 흐름 확인해보기

for i in range(5):
    print("반복이 되네 신기해!", i)
```

```python
#5. for문에서 range()함수 활용하기
# for i in range(10,4,-1):
#     print(i)

for i in reversed(range(10,4,-1)):
    print(i)
```

```python
#6. for문에서 시퀀스 객체로 반복하기
과일= ["두리안", "망고", "바나나", "멜론"]
# for i in 과일:
#   print(i)

for i in '닥스훈트':
    print(i, end='')
```

```python
#7. 예제로 for문 익혀보기
#문제 : 절대값 정수가 담긴 정수 변수와 그 정수의 부호를 차례로 담은 부호 변수가 있다
# 두 변수를 이용해 실제 정수의 합을 구하시오 (True: 양수, False:음수)

정수 = [10, 5, 8]
부호 = [True, False, True]

합 = 0
for k, v in zip(부호, 정수):
    if k:
        합 += v
    else:
        합 -= v
print(f'총 합은 {합}입니다.')
```
