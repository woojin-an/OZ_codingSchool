<span style="font-size:80%">[문제5]<br><br></span>
```Python 2버전대와 3버전대의 차이를 성명해주세요```

<span style="color:red">답변</span>

3버전의 기본적인 문자열은 Unicode로 훨씬 유용함 (문자,숫자,기호,이모티콘 모두 가능)

2버전은 ASC2


3버전이 정수 나눗셈에 더 유용함

3버전의 print 명령어 문법이 다름 (3버전은 괄호가 필요함, ex. print('A'))

3버전에서는 xrange를 사용하지 않음

3버전에서는 오류 발생 시 원인을 알려줌

3버전에서 __future__ 모듈을 사용하면 2버전 작업을 사용할 수 있음
