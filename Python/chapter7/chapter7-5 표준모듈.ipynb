[표준모듈]
모듈: 여러 변수와 함수를 가지고 있는 집합체 

파이썬에 기본적으로 내장되어있는 모듈을 '기본모듈' 이라고 함

입력방법: `import 모듈 이름`

[math 모듈]
* 수학과 관련된 기능을 가지고 있음

math 모듈 내 함수
* sin(x)
* cos(x)
* tan(x)
* ceil(x): 올림
* floor(x): 내림

```python
import math

print(math.sin(5))
print(math.cos(5))
print(math.ceil(3.6))
print(math.floor(4.1))
```

[from구문]

모듈안에 수많은 함수 중 우리가 원하는 함수만을 가져와 사용하면 문법적으로 굉장히 편리함



```python
from math import sin, cos, ceil, floor
print(sin(5))
print(cos(5))
print(ceil(3.6))
print(floor(4.1))
```

```python
# math모듈의 모든 함수를 가져오는 방법으로 import에 '*' 사용합니다.
from math import *
print(sin(5))
print(cos(5))
print(ceil(3.6))
print(floor(4.1))
```

[as구문]
모듈을 가져오다보면 이름이 충돌하는 경우가 있음

이럴때 모듈의 식별자를 만들어 사용할수있는 문법이 as구문임


```python
import math as m

print(m.sin(5))
print(m.cos(5))
print(m.ceil(3.6))
print(m.floor(4.1))
```

[datetime 모듈]
* 날짜와 시간 관련된 모듈

```python
import datetime

print("오늘 날짜와 시간 출력")
now = datetime.datetime.now()
print(now.year, '년')
print(now.month, '월')
print(now.day, '일')
print(now.hour, '시')
print(now.minute, '분')
print(now.second, '초')
```

```python
import datetime
print("다양한 시간 포맷")
output_1 = now.strftime("%Y.%m.%d %H:%M:%S")
output_2 = f'{now.year}년 {now.month}월 {now.day}일 {now.hour}시 {now.minute}분 {now.second}초'
output_3 = now.strftime("%Y{} %m{} %d{} %H{} %M{} %S{}").format(*"년월일시분초")
print(output_1)
print(output_2)
print(output_3)
print()
```

[time 모듈]
* 시간과 관련된 기능을 다루는 모듈
* 특정 시간동안 코드를 정지할 수 있는 기능이 있음

```python
import time
print("3초 뒤에 어떤 일이 일어날까?")
time.sleep(3)
print("3초가 지났습니다")
```
