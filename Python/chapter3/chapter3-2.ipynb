###list
* list란? 원소들이 연속적으로 저장되는 형태의 자료형 (시퀀스 자료형)
    * str(정복), list(정복 예정), tuple, dict, set
* 만드는 방법 - 대괄호로 묶어주고, 값들은 콤마로 구분

```python
# list

리스트 = []
print(리스트)

리스트 = [23, "이게 리스트다", 3,23, True, False]
print(리스트)
```

```python
리스트 = ["이", "렇", "게", "만", "들", "어", "요"]
리스트2 = list()

#range 연속된 숫자를 생산하는 기능
range(10)
list(range(0,10))
```

```python
#list(range(시작, 끝, 증가)) **인덱스의 개념이므로 주의해야함
리스트_1 = list(range(0,20,2))
print(리스트_1)

리스트_2 = list(range(20,0,-1))
print(리스트_2)
```

```python
# list 추가성질

#리스트를 이용해 변수에 값 넣기

a, b, c = ["들", "어", "가"]
print(a, b, c)

d, e, f = [7, 7, 8]

지갑 = [10000, 5000, 1000]
만원, 오천원, 천원 = 지갑
print(f'{만원} 2장, {오천원} 3장, {천원} 1장')
```

```python
# 슬라이싱

로또 = [3, 5, 15, 33, 41 ,44]
#로또 변수에 있는 리스트의 인덱스 1번부터 2번까지의 값을 출력
print(로또[1:3])
#로또 변수의 마지막 위치 출력
print(로또[-1])

#리스트 반대로 출력
print(로또[::-1])
#로또 리스트 안에 33이 있는지 확인
print(33 in 로또)
#로또 리스트를 로또2 리스트와 합치기
로또2= [2,12,15,24,33,39]
print(로또+로또2)

#range를 이용해서 1 부터 10 사이의 짝수만 들어있는 짝수(변수명) 리스트 만들기
짝수= list(range(2,12,2))
print(짝수)
#짝수 리스트에 들어있는 값을 2배 늘리고 다시 짝수 변수에 담기
짝수=짝수*2
print(짝수)
#짝수 리스트에 들어있는 요소의 갯수
print(len(짝수))
#짝수 리스트에 3번째 인덱스 출력
print(짝수[3])
#len()함수를 이용해 인덱스 마지막 값을 출력
print(짝수[len(짝수)-1])
```

```python

#list 값 넣고, 바꾸고, 지우고
mzfood = ["숙주", "분모자", "마라", "소세지", "소고기", "옥수수면"]

#값을 넣는 방법 2가지: append(), insert()
#append - 마지막 요소에 값을 넣어,
#insert - 원하는 곳에 값을 넣음

mzfood.append("고수")
print(mzfood)

mzfood.insert(0, "탕후루")
print(mzfood)

#값 지우기
#del

del mzfood[2]
print(mzfood)

#값 바꾸기
mzfood[-3] = "양고기"
print(mzfood)

#문자열 시퀀스
# 인사 = "반가워, 안녕, 헬로"
# 인사[2] = "웡"
# print(인사) #에러
```

```python
#
print(mzfood)

mzfood.reverse()
print(mzfood)

mzfood.sort()
print(mzfood)

```
