##시퀀스 자료형
* 문자열 :
    * 시퀀스형에서 사용할 수 있는 메소드(count(), index(), split() 등)
    * 슬라이싱 ([:],[::])
    * 요소들로 이루어져 가능했던 기능들 (요소합치기, 요소 찾기 등)
* list :
    * 시퀀스형에서 사용할 수 있는 메소드 (count(), index(), split() 등)
    * 슬라이싱 ([:],[::])
    * 요소들로 이루어져 가능했던 기능들 (요소합치기, 요소 찾기 등)
    * list 지원하는 다양한 메소드 (append(), insert(), del, sort(), reverse() 등)
* tuple :
    * 리스트처럼 요소를 일렬로 저장하지만 안에 저장된 요소를 변경, 추가, 삭제할 수 없다.
    * 값을 추가, 삭제, 변경할 수 없기 때문에 사용하지 못하는 메소드가 많음
    * tuple의 생성 방법: tuple= 값, 값,,, 또는 tuple = (값, 값 ,,,,)
    

```python
# 리스트 [], 튜플 ()
튜플 = (1, "이", 3.0)
print(type(튜플))
print(튜플)

# 리스트 = [1, "이", 3.0]
# print(type(리스트))
# print(리스트)

# 튜플2 = 1, "이", 3.0
# print(type(튜플2))
# print(튜플2)

#하나의 값만 들어있는 튜플만 만들고 싶다면?
튜플_솔로 = (1, )  #콤마로 구분하지 않으면 int 등으로 인식함
print(type(튜플_솔로))
print(튜플_솔로)

#값 추가, 삭제, 변경이 불가능?
print(튜플)
print(dir(튜플))
튜플_리스트화 = list(튜플)
print(type(튜플_리스트화))
```

```python
#번외
수강생 = "중꺽그마"

수강생_list = list(수강생)
수강생_tuple = tuple(수강생)

print(수강생_list)
print(수강생_tuple)

#join 함수 (흩어진 요소들을 문자열로 합쳐줌)
취뽀_수강생 = "".join(수강생_tuple)
print(type(취뽀_수강생))
print(취뽀_수강생)
```

```python
# 튜플과 리스트의 공통점
print(튜플)

#길이를 구할 수 있어요
len(튜플)
#곱하기 연산으로 요소를 복사할 수 있음
튜플 * 3
#튜플은 요소들끼리 더할 수 있다
튜플 + 튜플
#튜플은 인덱싱이 가능함
튜플[2]
#튜플의 인덱스
튜플.index(3.0)
#튜플 요소의 갯수
튜플.count(1)
#튜플도 슬라이싱이 가능
튜플[1:3]
```

#tuple 왜 쓸까?
(더 돋보일 수 있음)

1. 3go가 불가능함 ( 넣고, 삭제하고 ,바꾸고)
    여러명이 코딩을 하는 경우 변경되면 안되는 변수를 건드리는 경우를 방지
    값이 변경되면 안되는 중요한 데이터를 넣기 (보안적으로 관리하기 좋음 ex. 계좌, 비밀번호)
2. list는 tuple보다 무겁다
    단지 기능이 많아서 무거울까?
    파이썬은 동적배열을 지원하므로
    - list = [1,2,3,4] 생성하여 4개에 모두 값을 넣었음
    - 문제는 그 다음 값을 추가할 때 이미 공간이 다 찼으므로 추가할당을 진행해야 함 (파이썬에서 동적으로 메모리를 추가해야 하는데, 추가 시 메모리 크기는 더블링을 해야하므로 문제가 발생할 수 있음

3. list는 가변, tuple은 불변이다
    값이 고정되어야 하거나, 값이 작을 경우 튜플을 권장함

* 가변객체와, 불변객체
* 파이썬 리스트의 동적배열
