[문제1]
list와 tuple의 차이점은?

list는 가변형, tuple은 불변형 이다.

list는 객체 하나하나를 수정할 수 있지만,  
tuple의 경우 수정할 수 없어서 만약 필요한 경우 새로 할당하는 식의 다른 방법이 필요하다.

따라서 list는 수정이 필요할 수 있는 데이터에 쓰이며,  
tuple의 경우 수정되면 안되는 데이터 (예를 들면 비밀번호, 보안이 필요한 데이터 등)에 사용된다.

[문제3]  
Python의 list는 연결리스트(linked list) 인가요?

파이썬은 연결 리스트가 아닌, 내부적으로 동적 배열을 갖고있는 형태이다.
 * 동적 배열은 메모리를 할당받아 사용하고, 필요에 따라 크기를 조절하는 구조로써 이를 통해 요소끼리의 연산이 이루어 진다.
 * 연결 리스트는 말 그대로 리스트 사이에 연결이 된 것으로, 한 요소가 다른 요소를 가리키며 연결된 것이다. 이는 수정이 쉽지만 크기가 커질수록 데이터를 많이 할당할 수 있어 오래걸릴 수 있다.

[문제4]  
딕셔너리 자료형에 대해 설명해주세요

* 파이썬의 딕셔너리 자료형은 key값과 value값을 쌍으로 가지는 자료형이다.  
* 생성 방법은 {}(중괄호) 안에 키:값 형태로 작성하여 생성한다.  
    * 예) 손흥민 = {'키':183, '체중':78, '발사이즈':255, '혈액형':'AB'}  

* 여기서 키(key) 값은 고유하며(중복과 수정이 안됨), value값은 중복 가능하며 수정이 가능하다.
    * key를 수정하려면 삭제 후 다시 할당해야함
* list화 시켜서 데이터 접근이 가능하다.
    * 예) 손흥민_list = list(손흥민.items())
         print(손흥민_list[0][1])  
         **#출력 결과=>183**
