##### 함수와 메서드
>함수: 어디에 속해있지 않고 단독 모듈이라 함수를 그냥 호출하여 사용 가능함. 예) print(), input()
https://docs.python.org/ko/3/library/functions.html

>메소드: 함수의 클래스 안에 속해있고 클래스의 멤버 변수들을 이용해서 구현된 것
예) "까먹으신 건가요?" .count("까") / "까먹으신 건가요?" .index("까") 등

```python
인사 = input()
print(인사 + " 나도 신기해")
print(인사 + "    눈부시게")
```

```python
# 형변환 int(), float(), str(), bool()

정수 = int(input())
정수 = int(정수)
print(정수, type(정수))

실수 = float(input())
print(실수, type(실수))
```

그 외 입력값의 형변환이 가능한 종류
* tuple()
* list()
* dict()
* set()


```python
넌센스, 정답 = input().split("/")
print(f'넌센스 퀴즈 : {넌센스}')
print(f'정답 : {정답}')
```

```python
#시퀀스 자료형에 다 지원되는 메소드
번호1, 번호2, 번호3, 번호4 = input().split(",")

번호1 = int(번호1)
번호2 = int(번호2)
번호3 = int(번호3)
번호4 = int(번호4)

print(type(번호1), type(번호2), type(번호3), type(번호4))
```

```python
#map()
번호1, 번호2, 번호3, 번호4 = map(int, input().split(","))
print(type(번호1), type(번호2), type(번호3), type(번호4))
#split으로 구분하여 input한 값들을 int로 형변환 시킨 함수의 형태임 (위의 4줄을 요약한 느낌)
```

```python
#레거시 : 누군가 남겨두고간 코드
#리팩토링 : 결과의 변경 없이 코드의 구조를 재조정하는 것
#input

# fish__p, fish__s, fish__j = "붕어빵(팥)", "붕어빵(슈크림)", "붕어빵(잡채)"
fish__p, fish__s, fish__j = input('붕어빵(팥), 붕어빵(슈크림), 붕어빵(잡채)를 순서대로 개수를 입력하고 꼭 쉼표를 입력하고 안사는 붕어빵에는 0을 넣어줘').split(',')

fish__p, fish__s, fish__j = "붕어빵(팥)", "붕어빵(슈크림)", "붕어빵(잡채)"
fish__p__sell, fish__s__sell, fish__j__sell = 2000, 2500, 3000
order_s, order_j = 2, 3
order_s__pay, order_j__pay = fish__s__sell * order_s, fish__j__sell * order_j
total = order_s + order_j
total_pay = order_j__pay + order_s__pay

print(f'주문이 완료되었습니다. {fish__s} {order_s}개, {fish__j} {order_j}개 총 {total}개 결제 금액은 {total_pay}원 입니다 ')
```
