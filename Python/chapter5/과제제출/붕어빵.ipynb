```python
재고 = {
  "팥붕어빵" : 10,
  "슈크림붕어빵" : 8,
  "초코붕어빵" : 5
}

가격 = {
    "팥붕어빵" : 1000,
    "슈크림붕어빵" : 1200,
    "초코붕어빵" : 1500
}

판매량 = {
    "팥붕어빵" : 0,
    "슈크림붕어빵" : 0,
    "초코붕어빵" : 0
}
while True:
    모드 = input("모드를 선택하세요 (재고/주문/관리자/종료): ")
    #재고확인모드
    if 모드 == "재고":
        while True:
            현재재고 = input(f'재고를 알고싶은 항목을 입력하세요(팥붕어빵/슈크림붕어빵/초코붕어빵):')
            if 현재재고 in 재고:
                print(f"{현재재고}의 재고는 {재고[현재재고]}개 입니다.")
                
            else:
                print(f"'{현재재고}'는 유효하지 않은 항목입니다.")
            
            추가입력 = input("추가로 재고를 조회하시겠습니까? (예/아니오): ")
            if 추가입력.lower() != '예':
                break
    #주문모드
    elif 모드 == "주문":
        while True:
            주문종류 = input("주문할 붕어빵 종류를 입력하세요(팥붕어빵,슈크림붕어빵,초코붕어빵) 종료를 원하시면 '종료'를 입력하세요.:")
            if 주문종류 == "종료":
                break
            
            if 주문종류 in 재고:
                주문수량 = int(input("주문할 붕어빵 수량을 입력하세요:"))
            #재고 충분할 때
                if 재고[주문종류] >= 주문수량: #재고 수량 >= 주문수량 일 때
                    재고[주문종류] -= 주문수량 #재고수량에서 주문수량을 뺀걸 재할당
                    판매량[주문종류] += 주문수량
                    print(f"{주문종류} {주문수량}개 주문해주셔서 감사합니다!")
                
            #재고 부족할 때
                else:                           # 재고수량 < 주문수량 일 때
                    print(f"죄송합니다. 재고가 부족하여 판매할 수 없습니다.")
            else:
                print(f"'{주문종류}'는 유효하지 않은 항목입니다.")    
    
    #관리자모드
    elif 모드 == "관리자":
        while True:
            추가종류 = input("추가할 붕어빵 종류를 입력하세요(팥붕어빵,슈크림붕어빵,초코붕어빵). 종료를 원하시면 '종료'를 입력하세요.")
            if 추가종류 == "종료":
                break
            
            if 추가종류 in 재고:
                추가수량 = int(input(f"{추가종류}의 추가할 개수를 입력하세요"))
                재고[추가종류] += 추가수량
                print(f"{추가종류}의 재고가 {추가수량}개 추가되었습니다. 현재 재고는 {재고[추가종류]}개 입니다.")
            else:
                print(f"'{추가종류}'는 유효하지 않은 항목입니다.")
    
    #모드 종료
    elif 모드 == "종료":
        print("종료합니다.")
        print("최종 재고:", 재고)
        print("영업을 종료합니다.")
        총매출 = sum(판매량[상품] * 가격[상품] for 상품 in 판매량)
        print(f"총 매출: {총매출}원")
        print("모든 거래내역: ")
        break

    else:
        print("유효하지 않은 모드입니다. 다시 입력해주세요.")
        break
    
```
