오즈 (oz) 클래스에서 이름, 나이, 성별을 출력하는 who() 메소드를 추가하고 who()메소드를 이용해 아래 출력 결과를 만들어주세요


출력 결과 : 이름: 김코, 나이: 30, 코딩레벨: A

```python
class Oz:
	print("클래스 정복")
    
	def __init__(self, 이름, 나이, 코딩레벨):
		self.이름 = 이름
		self.나이 = 나이
		self.코딩레벨 = 코딩레벨

	def who(self):
		print(f"이름: {self.이름}, 나이: {self.나이}, 코딩레벨: {self.코딩레벨}")

coding = Oz("김코", 30, "A")
coding.who()
```

과제_1: BungeoppangShop 클래스 만들기

- __init__: 초기 재고, 가격, 판매량을 설정합니다.
- check_stock: 현재 재고를 출력합니다.
- process_order: 주문을 처리하고 재고를 업데이트합니다.
- admin_mode: 관리자 모드에서 재고를 추가합니다.
- calculate_total_sales: 총 매출을 계산하고 출력합니다.

```python
class 붕어빵:
    #초기재고, 가격, 판매량 설정
    def __init__(self,stock,price,sales):
        self.stock = {
            "팥붕어빵" : 10,
            "슈크림붕어빵" : 8,
            "초코붕어빵" : 5
        }
        self.price = {
            "팥붕어빵" : 1000,
            "슈크림붕어빵" : 1200,
            "초코붕어빵" : 1500
        }
        self.sales = {
            "팥붕어빵" : 0,
            "슈크림붕어빵" : 0,
            "초코붕어빵" : 0
        }
    
    #현재 재고를 출력
    def check_stock(self):
        print(f"{self.bread_type}의 재고는 {self.stock}개 입니다.")
		
    #주문을 처리하고 재고를 업데이트
    def process_order(self, bread_type, bread_count):
        if bread_count <= self.stock:
            self.stock -= bread_count
            self.sales += bread_count * self.price
            print(f"{bread_count}개의 {self.bread_type}을 주문하셨습니다. 남은 재고는 {self.stock}개입니다.")
        else:
            print("재고가 부족합니다.")

    #관리자 모드에서 재고를 추가
    def admin_mode(self, add_stock):
        self.stock += add.stock
        print(f"관리자 모드: {add_stock}개의 {self.bread_type}이 추가되었습니다. 현재 재고는 {self.stock}개입니다.")
    
    #총 매출을 계산하고 출력
    def calculate_total_sales(self):
        print(f"{self.bread_type} 붕어빵의 총 매출은 {self.sales}원입니다.")

def main():
    shop = 붕어빵()
    while True:
        mode = input("원하는 모드를 선택하세요(주문, 관리자, 종료): ")
        if mode == "종료":
            break
        if mode == "주문":
            continue
        if mode == "관리자":
            shop.admin_mode()
            continue

        
        
    if __name__ == "__main__":
        main()
        pass
```
