str : 문자열
특징 : 시퀀스 자료형  
시퀀스란? 요소들이 연속적으로 이어진 자료형  
값 하나하나를 요소라고 함.

```python
print(소속) #<--시퀀스
print(소속[0]) #<--요소들
print(소속[1]) #<--요소들
# [] 안의 숫자는 인덱스임
# 시퀀스의 인덱스를 넣었을때 요소들을 출력
```

```python
"매" in 소속
```

```python
생년월일 = "1994.05.02"
print(생년월일[0:4], 생년월일[5:7], 생년월일[8:10])
```

```python
생년월일 = "1994.05.02"
년 = 생년월일[0:4]
월 = 생년월일[5:7]
일 = 생년월일[8:10]
print(f'{년} {월} {일}')
```

```python
숫자 = "2131566115613213"
print(숫자[::])

#전체 출력 print(숫자[::])
#전체 거꾸로 출력
print(숫자[::-1])

#첫번째 인덱스 기준으로 2칸씩 띄어서 출력하기
print(숫자[::2])
```

```python
퀴즈 = "스님이 공중에 뜬다를 6글자로 말하면? 어중이떠중이"
print(퀴즈[-6:])
```
