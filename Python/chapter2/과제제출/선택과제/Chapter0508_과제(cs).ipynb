[문제3]

슬라이싱에 대해 간단히 설명해주세요

인덱싱은 시퀀스 중 어느 객체를 지정하는 반면, 슬라이싱은 원하는 부분의 객체를 추출하는 것이다.(또는 지정)  
시퀀스객체[x:y:z] 로 사용할 수 있는데,  
x 부터 y 까지 z 마다 객체를 추출하는 것이다.  
단, y번째 객체를 추출하는것이 아니라  
y번째 객체에서 slicing 하는 것 이므로
추출은 y-1 번째 객체
여기서 x를 생략할 경우 맨 처음(0) 으로 사용되고,  
y를 생략할 경우 맨 끝 인덱스로 인식하고,
z를 생략할 경우 증가폭을 설정하지 않은 것으로 인식한다.  
z 에 음수가 들어갈 경우 거꾸로 인덱스를 지정하게 된다.  
  


(예시)

a = "OZCODING SCHOOL"  
print(a[0:6])  
print(a[7:10])  
print(a[::])  
print(a[::-1])  
print(a[:3:])  
print(a[:10:3])  
print(a[::-2])
